## Under Sampling with NearMiss Method

# NearMiss

This procedures aims to select samples that are somewhat similar to the minority class, using 1 of three alternative procedures:

1) Select observations closer to the closest minority class

2) Select observations closer to the farthest minority class

3) Select observations furthest from their nearest neighbours

===

This procedure will select as many obserations from the majority class, as observations from the minority class are present in the dataset.

===

**Final Data size**: 2 x minority class

**Also, this under sampling method is recommeded to work better in text datasets where it consists of words and tags specially**

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# from sklearn
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# for metrics
from sklearn.metrics import roc_auc_score

# from imblearn
from imblearn.under_sampling import NearMiss
```

## create simulated data

```python
def create_data(sep):
    
    X,y = make_classification(n_samples=1000,
                              n_features=2,
                              n_redundant=0,
                              n_clusters_per_class=1,
                              weights=[0.99],
                              class_sep=sep,
                              random_state=1)
    
    X = pd.DataFrame(X, columns=['varA','varB'])
    y = pd.Series(y)
    
    return X,y
```

## resample using OneSidedSelection method

```python
# lets generate data when sep = 2
X,y = create_data(sep=2)
```

```python
X.shape, y.shape
```

```python
y.value_counts()
```

```python
nm = NearMiss(sampling_strategy='auto',          # removes from the majority class
                              n_neighbors=3,     # setting the nearest neighbours count ,
                              version=1,         # avg distance from nearest majority obs to closes minority obs
                              n_jobs=2)

X_res, y_res = nm.fit_resample(X,y)
print('Resample Data Shapes: ')
print(X_res.shape, y_res.shape)
print('Value Counts: ')
print(y_res.value_counts())
```

```python
# original data plot
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y)
plt.show()
```

```python
# under sampled data plot
sns.scatterplot(data=X_res, x = 'varA', y ='varB', hue=y_res)
plt.show()
```

- We can see this under sampling method reduces the majority class observation to twice that of the minorty ones
- Above, we can see the observations that are retained from majority ones are the ones that have the least average distance from the closest minority class observations
- Lets check for version 2 and 3 as well

```python
def check_version(version,X, y):
    nm = NearMiss(sampling_strategy='auto',          # removes from the majority class
                                  n_neighbors=3,     # setting the nearest neighbours count ,
                                  version=version,         # avg distance from nearest majority obs to closes minority obs
                                  n_jobs=2)

    X_res, y_res = nm.fit_resample(X,y)
    print('Resample Data Shapes: ')
    print(X_res.shape, y_res.shape)
    print('Value Counts: ')
    print(y_res.value_counts())
    
    # under sampled data plot
    sns.scatterplot(data=X_res, x = 'varA', y ='varB', hue=y_res)
    plt.title('For version {}'.format(version))
    plt.show()    
```

```python
# original data plot
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y)
plt.show()
```

```python
for version in [2,3]:
    check_version(version, X,y)
```

## Lets check for a partially separated class

```python
# original data plot
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y)
plt.show()

# we can very less separation and much overlapped since we provide a lower seperation value 
```

## Working on a real data

```python
# sampling to only 10000 observations to speed up the computation
df = pd.read_csv('kdd2004.csv').sample(10000)
df.head()
```

```python
df.shape
```

```python
# check the target values ratios
df['target'].value_counts()
```

```python
df['target'].value_counts()/len(df)
```

## split the data

```python
X_train, X_test, y_train,y_test = train_test_split(df.drop('target', axis = 1),
                                                  df['target'],
                                                  test_size=0.3,
                                                  random_state=0)
X_train.shape, X_test.shape
```

```python
# now lets prepare our under resampled data sets for version 1
nm = NearMiss(sampling_strategy='auto',          # removes from the majority class
                              n_neighbors=3,     # setting the nearest neighbours count ,
                              version=1,         # avg distance from nearest majority obs to closes minority obs
                              n_jobs=2)

X_nm1, y_nm1 = nm.fit_resample(X_train,y_train)
X_nm1.shape, y_nm1.shape
```

```python
# now lets prepare our under resampled data sets for version 2
nm2 = NearMiss(sampling_strategy='auto',          # removes from the majority class
                              n_neighbors=3,     # setting the nearest neighbours count ,
                              version=2,         # avg distance from nearest majority obs to closes minority obs
                              n_jobs=2)

X_nm2, y_nm2 = nm2.fit_resample(X_train,y_train)
X_nm2.shape, y_nm2.shape
```

```python
# now lets prepare our under resampled data sets for version 1
nm3 = NearMiss(sampling_strategy='auto',          # removes from the majority class
                              n_neighbors=3,     # setting the nearest neighbours count ,
                              version=3,         # avg distance from nearest majority obs to closes minority obs
                              n_jobs=2)

X_nm3, y_nm3 = nm3.fit_resample(X_train,y_train)
X_nm3.shape, y_nm3.shape
```

## Model performance on original and under sampled data

```python
def create_randomForest(X_train, X_test, y_train, y_test):
    rf = RandomForestClassifier(n_estimators=200,
                                n_jobs=2,
                                random_state=39,
                                max_depth=4)
    
    rf.fit(X_train, y_train)
    probs_train = rf.predict_proba(X_train)
    probs_test = rf.predict_proba(X_test)
    
    # get metrics
    print('roc auc score for train data: {}'.format(roc_auc_score(y_train, probs_train[:,1])))
    print('roc auc score for test data: {}'.format(roc_auc_score(y_test, probs_test[:,1])))
```

```python
# for Original DataSet
create_randomForest(X_train, X_test, y_train, y_test)
```

```python
# for condensed dataset
create_randomForest(X_nm1, X_test, y_nm1, y_test)
```

```python
# for condensed dataset
create_randomForest(X_nm2, X_test, y_nm2, y_test)
```

```python
# for condensed dataset
create_randomForest(X_nm3, X_test, y_nm3, y_test)
```

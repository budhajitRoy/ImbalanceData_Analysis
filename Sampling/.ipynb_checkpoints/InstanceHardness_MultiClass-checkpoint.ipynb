## Under Sampling using Instance Hardness Method for multi class target datasets

# Instance Hardness Threshold

Instance hardness is the probability of an observation being miss-classified. So in other words, the instance hardness is 1 minus the probability of the class. For class 1:

- **instance hardness = 1 - p(1)**

The instance hardness depends on the algorithm used in the classification and in class separability.

## Instance Hardness Filtering

Instance Hardness filtering consists in determining the instance hardness and removing the observations that are harder to classify. In other words, removing the observations with high **instance hardness** or low **class probability**. The question is, how to determine the threshold.

In this notebook we will:

- Determine the instance hardness of observations
- Filter observations from the majority class based of an arbitrary threshold
- Filter observations from the majority class as proposed by Imbalanced-learn
- Use imbalance-learn to implement Instance Hardness

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# from sklearn
from sklearn.model_selection import (
    train_test_split,
    StratifiedKFold,
    cross_val_predict
)

from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# one Vs Rest Classifier for multiclass
from sklearn.multiclass import OneVsRestClassifier

# for metrics
from sklearn.metrics import roc_auc_score

# from imblearn
from imblearn.under_sampling import InstanceHardnessThreshold
```

## create simulated data

```python
# function to create simulated data
# the user can specify the number of classes
# the imbalance, and how well separated the classes are
def create_data(weights,sep,n_classes):
    
    X,y = make_classification(n_samples=1000,
                              n_features=2,
                              n_redundant=0,
                              n_classes=n_classes,
                              n_clusters_per_class=1,
                              weights=weights,
                              class_sep=sep, # how separate the classes are
                              random_state=1)
    
    X = pd.DataFrame(X, columns=['varA','varB'])
    y = pd.Series(y)
    
    return X,y
```

```python
# create a dataset and try plotting

X,y = create_data(weights=[0.5,0.4, 0.1], sep=0.5, n_classes=3)

# check the data 
print('Data Shapes:')
print(X.shape, y.shape)

# check the balance ratio
print('Balance Ratio:')
print(y.value_counts())

# plot
sns.scatterplot(data = X, x = 'varA', y ='varB', hue=y)
plt.title('Multiclass Classification')
plt.show()
    
```

- We can see some overlap in the two classes

## Train using Random Forest
We will use Random Forests to determine instance hardness. We will do so with cross-validation, to try and perform the procedure as closely as possible to the imbalanced-learn implementation.
To determine instance hardness, we will do in a 1 vs Rest manner. So we wrap the algorithm with a OneVsRestClassifier.

More details on the sklearn functions we use:

- [StratifiedKFold](https://scikit-learn.org/0.16/modules/generated/sklearn.cross_validation.StratifiedKFold.html)
- [cross_val_predict](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_predict.html)

```python
# setting up Stratified Cross validation scheme
skf = StratifiedKFold(n_splits=3,   # 3 Fold Cross validation
                      shuffle=True,
                      random_state=1)

skf
```

```python
# set up Random Forest Classifier and wrapping it using OneVsRestClassifier
rf = OneVsRestClassifier(
    RandomForestClassifier(n_estimators=10, max_depth=2, random_state=1),
    n_jobs=2
)
```

## Calculating Instance Hardness value

```python
# getting the predictions from the Random Forest with cross validtion
probs = cross_val_predict(rf,
                          X,
                          y,
                          cv=skf,
                          n_jobs=2,
                          method='predict_proba')
probs
# this returns an array of probabilities w.r.t to class 0 and class 1
```

```python
# lets convert this probs into a DataFrame
probs = pd.DataFrame(probs, columns = ['p(0)','p(1)','p(2)'])
probs
```

The class 0 and class 1 here is the majority class. So we want to under-sample observations from these two class. We can do so by removing observations with low p(0) and p(1), which is the same as saying high instance hardness. Remember that instance hardness is 1 - the probability of the class.

## Under-sample observations from the majority class

### Arbitrary threshold

```python
# checking the probs statistics to pick the arbitrary threshold
probs.describe()
```

We see that most observations of the majority classes 0 and 1 show have probs greater than 0.26 (given by the 25 percentile). We can try and remove observations under 0.30. This is arbitrary of course, we would have to try several thresholds and see what works best for our data.

```python
y.value_counts()
```

```python
# now we can remove either
# those observations whose probability is less than 0.30
# or whose instance hardness is 1 - prob = 1 - 0.30
# these both are the same

# so, lets remove those observations only from the majority classes (class 0 and 1) 
# whose prob returned by RF is less 0.30

condition_0 = (y == 0) & (probs['p(0)'] <= 0.30)
condition_1 = (y == 1) & (probs['p(1)'] <= 0.30)

# getting the filtered data from X
filtered_0 = X.loc[condition_0]
filtered_1 = X.loc[condition_1]
filtered_0.shape, filtered_1.shape
```

```python
# original X shape
X.shape
```

Lets identify first which observations will be removed from class 0

```python
# lets identify which observations will be removed
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y + 3*condition_0)
plt.show()
```

The darker colored observations are the ones that will be removed from class 0

```python
# Similarly lets identify which observations will be removed from class 1
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y + 2*condition_1)
plt.show()
```

```python
# dropping the hard instances
X_resampled = X.drop(filtered_0.index)
X_resampled = X_resampled.drop(filtered_1.index)

y_resampled = y.drop(filtered_0.index)
y_resampled = y_resampled.drop(filtered_1.index)

X_resampled.shape, y_resampled.shape
```

So, 83 observations are removed

```python
# target balance after filtering
y_resampled.value_counts()
```

```python
# plot filtered data

sns.scatterplot(data=X_resampled, x = 'varA', y ='varB', hue=y_resampled)
plt.title('Resampled Data Distribution')
plt.show()
```

## Under-sample data to match balancing ratio

- So far we were using an arbitraty threshold value by checking the statistics of the probability distribution
- Now lets try to find a threshold that equilibrates the balancing ratio.

```python
# first lets check the target value counts and balance ratios
balance = y.value_counts()
balance
```

```python
(balance[0] + balance[1])/balance[2]
```

- We can see for an observation from class 2, there are around 8.5 observations from class 0 and class 1 combined

```python
# when class = 0
class_ = 0

# number of samples needed given by the minority class samples
n_samples = balance[2]

# total number of samples, given by the class being considered i.e. class 0
n_total_samples = balance[class_]

# we will calculate the percentile on which we find these number of observations
perc = n_samples/n_total_samples
perc
```

```python
# we want the observations with the highest p(0)
# so we are looking at the top percentage and not at the bottom percentage

# thus we need to find the top percentile which is 1 - perc

perc = (1 - perc) * 100

perc
```

```python
# lets find the threshold at which we find 88% of the observations
threshold = np.percentile(probs[y==class_]['p(0)'], perc)
threshold
```

```python
# lets visually check this threshold value
probs[y==class_]['p(0)'].hist(cumulative=True, bins=100)
plt.axvline(x=threshold, c = 'red')
```

```python
# now we will remove observations from class 0
# where the probability is below the threshold

cond_0 = (y==class_) & (probs['p(0)'] < threshold)

filtered_0 = X.loc[cond_0]
filtered_0.shape
```

```python
# now lets plot and check visually
sns.scatterplot(data= X, x ='varA', y ='varB', hue = y + 3*cond_0)
```

- Now the darker colored ones represented by 3 are the observations from class 0 that will be remove
- Lets repeat the same procedure for class 1

```python
class_ = 1

n_samples = balance[2]

total_samples = balance[class_]

perc_1 = n_samples/total_samples

# calculate the lower percentage value
perc_1 = (1 - perc_1) * 100
perc_1
```

```python
# calculate threshold

threshold_1 = np.percentile(probs[y == class_]['p(1)'], perc_1)
threshold_1
```

```python
# now lets remove the observations from class 1 which are below this threshold value

cond_1 = (y == class_) & (probs['p(1)'] < threshold_1)

filtered_1 = X.loc[cond_1]
filtered_1.shape
```

```python
# check the observations visually

sns.scatterplot(data = X, x = 'varA', y= 'varB', hue = y + 2*cond_1)
```

```python
# removing these filtered observations from our data
X_resampled = X.drop(filtered_0.index)
X_resampled = X_resampled.drop(filtered_1.index)

y_resampled = y.drop(filtered_0.index)
y_resampled = y_resampled.drop(filtered_1.index)
```

```python
X_resampled.shape, y_resampled.shape
```

```python
# check the target balance after filtering
y_resampled.value_counts()
```

As we can see, we did reduce the number of observations from the majority, but we did not achieve the desired balance, because there are many observations that have a similar probability value. So this algorithm retains them all.

```python
# plot the resampled data
sns.scatterplot(data = X_resampled, x ='varA', y ='varB', hue=y_resampled)
```

### Instance hardness with imbalanced-learn

Let's repeat the exercise but now using imbalanced-learn

[InstanceHardnessThreshold](https://imbalanced-learn.org/stable/references/generated/imblearn.under_sampling.InstanceHardnessThreshold.html)

```python
# setting up instance Hardness threshold

iht = InstanceHardnessThreshold(estimator=rf,
                          sampling_strategy='auto', # to remove from the majority class
                          cv=3, # 3-fold cross validation
                          n_jobs=2,
                          random_state=1)
X_resampled, y_resampled = iht.fit_resample(X,y)
X_resampled.shape, y_resampled.shape
```

```python
# Original shape of the data
X.shape, y.shape
```

```python
# balancing ratio
y_resampled.value_counts()
```

```python
# original data plot 
sns.scatterplot(data= X, x='varA', y='varB', hue=y)
```

```python
# sampled data plot
sns.scatterplot(data= X_resampled, x='varA', y='varB', hue=y_resampled)
```

```python
# indexes of the samples that are retain
iht.sample_indices_
```

```python
# let's find the removed observations

removed_obs = [0 if x in iht.sample_indices_ else 1 for x in X.index ]

len(removed_obs), np.sum(removed_obs)
```

- So, 233 observations are removed

```python
#plot the original data highlighting removed observations

sns.scatterplot(data= X, x ='varA', y='varB', hue = y + 2*pd.Series(removed_obs))
```

- So, we can see good amout of data are being removed and we kind of have a balance data now to work with
- But still there is always a trade off.. removing data means, we can still have the chance to remove observations that can be critical to our model prediction. So these things always has to be considered.

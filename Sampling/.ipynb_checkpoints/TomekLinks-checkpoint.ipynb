# Condensed Nearest Neighbours (CNN)

The algorithms works as follows:

1) Put all minority class observations in a group, typically group O

2) Add 1 sample (at random) from the majority class to group O

3) Train a KNN with group O

4) Take a sample of the majority class that is not in group O yet

5) Predict its class with the KNN from point 3

6) If the prediction was correct (meaning it belongs to the majority class), remove this observation, go to 4 and repeat

7) If the prediction was incorrect, add that sample to group O, go to 3 and repeat

8) Continue until all samples of the majority class were either assigned to O or left out

9) Final version of Group O is our undersampled dataset


====

- **Criteria for data exclusion**: Samples outside the boundary between the classes
- **Final Dataset size**: varies

====

This algorithm tends to pick points near the fuzzy boundary between the classes, and transfer those to the group O, in our example. 

If the classes are similar, group O will contain a fair amount of both classes. If the classes are very different, group O would contain mostly 1 class, the minority class.

**Caution:**

- CNN tends to add noise to the undersampled dataset
- Computationally expensive, because it trains 1 KNN every time an observation is added to the minority class group.

In this notebook, we will first understand what Condensed Nearest Neigbours is doing with simulated data, and then compare its effect on model performance with real data.

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# from sklearn
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# for metrics
from sklearn.metrics import roc_auc_score

# from imblearn
from imblearn.under_sampling import CondensedNearestNeighbour
```

## create simulated data

```python
def create_data(sep):
    
    X,y = make_classification(n_samples=1000,
                              n_features=2,
                              n_redundant=0,
                              n_clusters_per_class=1,
                              weights=[0.99],
                              class_sep=sep,
                              random_state=0)
    
    X = pd.DataFrame(X, columns=['varA','varB'])
    y = pd.Series(y)
    
    return X,y
```

```python
# create data for different sep values and plot them
for i in [0,1,2]:
    X,y = create_data(i)
    
    #df = pd.DataFrame(pd.concat([X,y], axis = 1), columns = ['varA','varB','target'])
    
    # plot
    sns.scatterplot(data = X, x = 'varA', y ='varB', hue=y)
    plt.title('Separation value: {}'.format(i))
    plt.show()
    
```

```python
# original data shape
X.shape, y.shape
```

```python
# original data value counts
y.value_counts()
```

## resample using CondensedNearestNeighbour method

```python
# lets generate data when sep = 2
X,y = create_data(sep=2)
```

```python
cond = CondensedNearestNeighbour(sampling_strategy='auto', # this will reduce the majority class by default
                                 n_neighbors=1, # knn = 1 default value
                                 n_jobs=2,
                                 random_state=0)

X_res, y_res = cond.fit_resample(X,y)
print('Resample Data Shapes: ')
print(X_res.shape, y_res.shape)
print('Value Counts: ')
print(y_res.value_counts())
```

```python
# original data plot
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y)
plt.show()
```

```python
# condensed data plot
sns.scatterplot(data=X_res, x = 'varA', y ='varB', hue=y_res)
plt.show()
```

Condensed Nearest Neighbours retains the observations from the majority class, that are more similar to those in the minority class.

## Lets check for a partially separated class

```python
X,y = create_data(sep=0.5)
```

```python
# create resampled data for the partially separated class dataset

cond = CondensedNearestNeighbour(sampling_strategy='auto', # this will reduce the majority class by default
                                 n_neighbors=1, # knn = 1 default value
                                 n_jobs=2,
                                 random_state=0)

X_res, y_res = cond.fit_resample(X,y)
print('Resample Data Shapes: ')
print(X_res.shape, y_res.shape)
print('Value Counts: ')
print(y_res.value_counts())
```

```python
# original data plot
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y)
plt.show()

# we can very less separation and much overlapped since we provide a lower seperation value 
```

```python
# condensed data plot
sns.scatterplot(data=X_res, x = 'varA', y ='varB', hue=y_res)
plt.show()
```

- Similarly, here also we can see that observations from majority that are closed to minority ones are being preserved

## Different KNN values

```python
def create_resampled_data(knn,X,y):
    cond = CondensedNearestNeighbour(sampling_strategy='auto', # this will reduce the majority class by default
                                     n_neighbors=knn, # knn = 1 default value
                                     n_jobs=2,
                                     random_state=0)

    X_res, y_res = cond.fit_resample(X,y)
    print('Resample Data Shapes for knn = {}'.format(knn))
    print(X_res.shape, y_res.shape)
    print('Value Counts: ')
    print(y_res.value_counts())
```

```python
# when sep = 2
X, y = create_data(sep=2)

# original data plot
sns.scatterplot(data=X, x = 'varA', y ='varB', hue=y)
plt.title('Original Data plot')
plt.show()

for i in range(1,5):
    create_resampled_data(i,X,y)    
    # condensed data plot
    sns.scatterplot(data=X_res, x = 'varA', y ='varB', hue=y_res)
    plt.title('Condensed when knn = {}'.format(i))
    plt.show()
    
```

- Here we can see from above plots that the further observations are being removed and the similar observations from both the majority and minority classes are preserved
- The original data is spread out, but in the condensed data, we can see the value ranges are upto 2 for varA and upto 2.5 for varB

## Working on a real data

```python
# sampling to only 10000 observations to speed up the computation
df = pd.read_csv('kdd2004.csv').sample(10000)
df.head()
```

```python
df.shape
```

```python
# check the target values ratios
df['target'].value_counts()
```

```python
df['target'].value_counts()/len(df)
```

## split the data

```python
X_train, X_test, y_train,y_test = train_test_split(df.drop('target', axis = 1),
                                                  df['target'],
                                                  test_size=0.3,
                                                  random_state=0)
X_train.shape, X_test.shape
```

```python
# now lets prepare our condensed under resampled data sets
cond = CondensedNearestNeighbour(sampling_strategy='auto',
                                n_neighbors=1,
                                n_jobs=2,
                                random_state=0)

X_cond, y_cond = cond.fit_resample(X_train, y_train)
X_cond.shape, y_cond.shape
```

```python
# target value ratio
y_cond.value_counts()
```

## Model performance on original and condensed resample data

```python
def create_randomForest(X_train, X_test, y_train, y_test):
    rf = RandomForestClassifier(n_estimators=200,
                                n_jobs=2,
                                random_state=39,
                                max_depth=4)
    
    rf.fit(X_train, y_train)
    probs_train = rf.predict_proba(X_train)
    probs_test = rf.predict_proba(X_test)
    
    # get metrics
    print('roc auc score for train data: {}'.format(roc_auc_score(y_train, probs_train[:,1])))
    print('roc auc score for test data: {}'.format(roc_auc_score(y_test, probs_test[:,1])))
```

```python
# for Original DataSet
create_randomForest(X_train, X_test, y_train, y_test)
```

```python
# for condensed dataset
create_randomForest(X_cond, X_test, y_cond, y_test)
```

## On SVM classifier

```python
def create_svc(X_train, X_test, y_train, y_test):
    
    svc = SVC(max_iter=1000, random_state=0, probability=True)
    
    svc.fit(X_train, y_train)
    probs_train = svc.predict_proba(X_train)
    probs_test = svc.predict_proba(X_test)
    
    # get metrics
    print('roc auc score for train data: {}'.format(roc_auc_score(y_train, probs_train[:,1])))
    print('roc auc score for test data: {}'.format(roc_auc_score(y_test, probs_test[:,1])))
```

```python
# for original data
create_svc(X_train, X_test, y_train, y_test)
```

```python
# for condensed dataset
create_svc(X_cond, X_test, y_cond, y_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
def create_log(X_train, X_test, y_train, y_test):
    
    log = LogisticRegressionCV(max_iter=10000, random_state=0, solver='saga', n_jobs=2)
    
    log.fit(X_train, y_train)
    probs_train = log.predict_proba(X_train)
    probs_test = log.predict_proba(X_test)
    
    # get metrics
    print('roc auc score for train data: {}'.format(roc_auc_score(y_train, probs_train[:,1])))
    print('roc auc score for test data: {}'.format(roc_auc_score(y_test, probs_test[:,1])))
```

```python
# for original data
create_log(X_train, X_test, y_train, y_test)
```

```python
# for condensed dataset
create_log(X_cond, X_test, y_cond, y_test)
```

- For logistic regression model, we can see the model performs much better with the Condensed Undersample data
- Lets try by under sampling with various knn values and check the model comparison

```python
# create a function to create condensed under sampled data and check the metrics

def create_condensed_data(knn,X_train, X_test,y_train,y_test):
    cond = CondensedNearestNeighbour(sampling_strategy='auto', # this will reduce the majority class by default
                                     n_neighbors=knn, # knn = 1 default value
                                     n_jobs=2,
                                     random_state=0)

    X_res, y_res = cond.fit_resample(X_train,y_train)
    # for original data
    print("*******************")
    print('Metrics for Original Data: {}'.format(knn))
    create_log(X_train, X_test, y_train, y_test)
    # for condensed data
    print('Metrics for Condensed Data: {}'.format(knn))
    create_log(X_res, X_test, y_res, y_test)
```

```python
# trying for knn = 1,2 and 3
for knn in range(1,4):
    create_condensed_data(knn,X_train, X_test,y_train,y_test)
```

- Here, we can see that our Logistic Model definitely performs better in the condensed under-sampled data than the original data
- For different values of knn, we do not see much difference in the performance metrics
- Maybe, we the one with knn = 1, I think performed little better than the rest

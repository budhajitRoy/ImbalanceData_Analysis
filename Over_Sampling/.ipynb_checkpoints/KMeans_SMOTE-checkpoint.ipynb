# SVM-SMOTE

Creates new samples by interpolation of samples of the support vectors from minority class and its closest neighbours.

```python
# import libraries

import numpy as np
import pandas as pd

# for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# from sklearn 
from sklearn.datasets import make_blobs
from sklearn import svm

# from imblearn
from imblearn.over_sampling import SVMSMOTE
```

```python
# creating toy dataset

X, y = make_blobs(n_samples=1600, n_features=2, centers=[(0,0),(5,5)], random_state= 42,  cluster_std=1.5)

X.shape, y.shape
```

```python
# creating a data frame 

X = pd.DataFrame(data = X, columns = ['varA', 'varB'])
y = pd.Series(y)

# now, under sample the data for y ==1 
X = pd.concat([X[y == 0], X[y==1].sample(200, random_state=42)], axis = 0)

y = y.loc[X.index]

# display the shape
X.shape, y.shape
```

```python
# check the value counts for the target class
y.value_counts()
```

```python
# plot the original data as per the target class

sns.scatterplot(data = X, x = 'varA', y = 'varB', hue = y, alpha=0.5)
plt.title('Original data')
plt.show()
```

## Fitting a SVM 

```python
clf = svm.SVC(kernel='linear')
clf.fit(X,y)
```

```python
# can find the support vectors
# basically this gives the indices of the support vector observations
clf.support_
```

```python
# count of total support vectors
len(clf.support_)
```

```python
# get the support vectors - their co-ordinates
clf.support_vectors_
```

```python
clf.support_vectors_[:, 0]
```

```python
clf.support_vectors_[:, 1]
```

```python
# plotting the support vector points

sns.scatterplot(data = X, x = 'varA', y = 'varB', hue = y, alpha = 0.5)
plt.title('Support Vector points')

plt.scatter(clf.support_vectors_[:, 0],clf.support_vectors_[:,1],
            s = 100,
            edgecolors='black',
            linewidth=1,
            facecolors='none')

plt.show()
```

```python
# plotting the decision boundaries

sns.scatterplot(data = X, x = 'varA', y = 'varB', hue = y, alpha = 0.5)
plt.title('Decision Boundaries')

plt.show()
```

```python
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# add decision boundaries

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                    np.linspace(ylim[0], ylim[1], 50))
```

```python
# decision_function
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], clf.coef_[0]) + clf.intercept_[0]

print(Z)
Z = Z.reshape(xx.shape)
print('Value of Z')
print(Z)
```

```python
# plot the decision boundaries
sns.scatterplot(
    data=X, x="varA", y="varB", hue=y, alpha=0.5
)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# add decision boundaries
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))

# decision_function
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], clf.coef_[0]) + clf.intercept_[0]
Z = Z.reshape(xx.shape)
    
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
                linestyles=['--', '-', '--'])
plt.show()
```

```python
## adding the support vector points

# plot the decision boundaries
sns.scatterplot(
    data=X, x="varA", y="varB", hue=y, alpha=0.5
)


plt.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1] , s = 100,
           edgecolors='black', facecolors = 'none')

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# add decision boundaries
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))

# decision_function
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], clf.coef_[0]) + clf.intercept_[0]
Z = Z.reshape(xx.shape)
    
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
                linestyles=['--', '-', '--'])
plt.show()
```

## SVM SMOTE

[SVMSMOTE](https://imbalanced-learn.org/stable/references/generated/imblearn.over_sampling.SVMSMOTE.html)

```python
# create SVM SMOTE
from sklearn import svm
sv_smote = SVMSMOTE(sampling_strategy='auto', # samples from the minority class-- this is default
                        random_state=0,
                     svm_estimator=svm.SVC(kernel ='linear'), # setting up the svm classifier to get the support vectors
             k_neighbors=5,               # neighbours to generate new samples or synthetic samples
                     m_neighbors=10,        # to decide whether to perform interpolation or exra polation
             n_jobs=2) 

X_resampled, y_resampled = sv_smote.fit_resample(X,y)
X_resampled.shape, y_resampled.shape
```

```python
# original data shape
X.shape, y.shape
```

```python
# balance values
# original data
y.value_counts()
```

```python
y_resampled.value_counts()
```

```python
# plot the original data

sns.scatterplot(data = X, x = 'varA', y = 'varB', hue = y, alpha=0.5)
plt.title('Original data')
plt.show()
```

```python
# plot the resampled data - Borderline 1

sns.scatterplot(data = X_resampled, x = 'varA', y = 'varB', hue = y_resampled, alpha=0.5)
plt.title('Resampled data')
plt.show()
```

```python
## adding the support vector points

# plot the decision boundaries
sns.scatterplot(
    data=X_resampled, x="varA", y="varB", hue=y_resampled, alpha=0.5
)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# add decision boundaries
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))

# decision_function
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], clf.coef_[0]) + clf.intercept_[0]
Z = Z.reshape(xx.shape)
    
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
                linestyles=['--', '-', '--'])
plt.show()
```

We can see that most of the synthetic examples were created closer to the support vectors from the minority class, within the decision boundary from the SVM.
